### 랜덤 포레스트

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

X = telecom_df.drop(['telecom_change_yn'], axis=1)
y = telecom_df['telecom_change_yn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)

rf_clf = RandomForestClassifier(max_depth=7, random_state=24)

rf_clf.fit(X_train, y_train)

print(rf_clf.score(X_train, y_train), rf_clf.score(X_test, y_test))

feat_impt_ser = pd.Series(rf_clf.feature_importances_, index=X.columns).sort_values(ascending=False)
# print(feat_impt_ser)

rf_pred = rf_clf.predict(X_test)
print(f"랜덤포레스트 정확도: {accuracy_score(y_test, rf_pred):.4f}")
print(classification_report(y_test, rf_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

telecom_df = pd.read_csv('data/processed/telecom2_encoding.csv')

X = telecom_df.drop(['telcom_change_yn', 'skt_change', 'kt_change', 'lgu_change', 'mvno_change'], axis=1)
y = telecom_df['telcom_change_yn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)

rf_clf = RandomForestClassifier(max_depth=7, random_state=24)

rf_clf.fit(X_train, y_train)

print(rf_clf.score(X_train, y_train), rf_clf.score(X_test, y_test))

feat_impt_ser = pd.Series(rf_clf.feature_importances_, index=X.columns).sort_values(ascending=False)
# print(feat_impt_ser)

rf_pred = rf_clf.predict(X_test)
print(f"랜덤포레스트 정확도: {accuracy_score(y_test, rf_pred):.4f}")
print(classification_report(y_test, rf_pred))
```

```python
from xgboost import XGBClassifier

telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

X = telecom_df.drop(['telecom_change_yn'], axis=1)
y = telecom_df['telecom_change_yn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)

xgb_clf = XGBClassifier(
    n_estimators=200,     
    learning_rate=0.05,    
    max_depth=10,           
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

xgb_clf.fit(X_train, y_train)

print(xgb_clf.score(X_train, y_train), xgb_clf.score(X_test, y_test))

feat_impt_ser = pd.Series(xgb_clf.feature_importances_, index=X.columns).sort_values(ascending=False)
# print(feat_impt_ser)

xgb_pred = xgb_clf.predict(X_test)
print(f"xgb 정확도: {accuracy_score(y_test, xgb_pred):.4f}")
print(classification_report(y_test, xgb_pred))
```
