### 랜덤 포레스트

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

telecom_df['usage_income_ratio'] = (
    telecom_df['phone_usage_per_m'] / (telecom_df['income'] + 1)
)

X = telecom_df.drop(['telecom_change_yn'], axis=1)
y = telecom_df['telecom_change_yn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)

rf_clf = RandomForestClassifier(max_depth=7, random_state=24)

rf_clf.fit(X_train, y_train)

print(rf_clf.score(X_train, y_train), rf_clf.score(X_test, y_test))

feat_impt_ser = pd.Series(rf_clf.feature_importances_, index=X.columns).sort_values(ascending=False)
# print(feat_impt_ser)

rf_pred = rf_clf.predict(X_test)
print(f"랜덤포레스트 정확도: {accuracy_score(y_test, rf_pred):.4f}")
print(classification_report(y_test, rf_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

# 소득 대비 핸드폰 요금비 부담률
telecom_df['usage_income_ratio'] = (
    telecom_df['phone_usage_per_m'] / (telecom_df['income'] + 1)
)

# 직업 기반 소득 안전성
telecom_df['income_if_job'] = (
    telecom_df['job'] * telecom_df['income']
)

X = telecom_df.drop(['telecom_change_yn', 'mar_1', 'mar_2', 'mar_3', 'mar_4','gender', 'income', 'phone_usage_per_m', 'job', 'id'], axis=1)
y = telecom_df['telecom_change_yn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)

rf_clf = RandomForestClassifier(max_depth=7, random_state=24)

rf_clf.fit(X_train, y_train)

print(rf_clf.score(X_train, y_train), rf_clf.score(X_test, y_test))

feat_impt_ser = pd.Series(rf_clf.feature_importances_, index=X.columns).sort_values(ascending=False)
# print(feat_impt_ser)

rf_pred = rf_clf.predict(X_test)
print(f"랜덤포레스트 정확도: {accuracy_score(y_test, rf_pred):.4f}")
print(classification_report(y_test, rf_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier

telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

X = telecom_df.drop(['telecom_change_yn', 'mar_1', 'mar_2', 'mar_3', 'mar_4','gender', 'id'], axis=1)
y = telecom_df['telecom_change_yn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)


xgb_clf = XGBClassifier(
    n_estimators=500,
    learning_rate=0.03,
    max_depth=5,             
    min_child_weight=3,
    scale_pos_weight=1.5,     
    random_state=42,
    eval_metric='logloss'
)

xgb_clf.fit(X_train, y_train)


# 모델에서 각 데이터의 예측 확률값만 추출
xgb_proba = xgb_clf.predict_proba(X_test)[:, 1]

# 임계값(Threshold)을 0.47로 설정
custom_pred = (xgb_proba > 0.47).astype(int)

print(X.columns.tolist())

print(classification_report(y_test, custom_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE


telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

# 소득 대비 핸드폰 요금비 부담률
telecom_df['usage_income_ratio'] = (
    telecom_df['phone_usage_per_m'] / (telecom_df['income'] + 1)
)

# 직업 기반 소득 안전성
telecom_df['income_if_job'] = (
    telecom_df['job'] * telecom_df['income']
)

X = telecom_df.drop(['telecom_change_yn', 'mar_1', 'mar_2', 'mar_3', 'mar_4','gender', 'income', 'phone_usage_per_m', 'job', 'id'], axis=1)
y = telecom_df['telecom_change_yn']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# 3. SMOTE 적용
smote = SMOTE(random_state=42)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)

print(f"--- 데이터 분포 확인 ---")
print(f"SMOTE 적용 전 훈련셋: {X_train.shape}")
print(f"SMOTE 적용 후 훈련셋: {X_train_over.shape}")
print(f"레이블 분포(1이 늘어났는지 확인): \n{pd.Series(y_train_over).value_counts()}")
print("-" * 30)


xgb_clf = XGBClassifier(
    n_estimators=1000,
    learning_rate=0.03,
    max_depth=7,          
    min_child_weight=3,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)


xgb_clf.fit(X_train_over, y_train_over)


train_score = xgb_clf.score(X_train_over, y_train_over)
test_score = xgb_clf.score(X_test, y_test)

print(f"훈련 점수(오버샘플링): {train_score:.4f}")
print(f"테스트 점수(실제): {test_score:.4f}")
print("-" * 30)

# 6. 상세 리포트
xgb_pred = xgb_clf.predict(X_test)
print(f"최종 xgb 정확도: {accuracy_score(y_test, xgb_pred):.4f}")
print("\n[상세 분류 리포트]")
print(classification_report(y_test, xgb_pred))

# 7. 피처 중요도 확인 (상위 10개)
feat_impt_ser = pd.Series(xgb_clf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\n[중요 피처 상위 10개]")
print(feat_impt_ser.head(10))
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree


telecom_df = pd.read_csv('../../data/processed/telecom_encoding.csv')

X = telecom_df.drop(['telecom_change_yn'], axis=1)
y = telecom_df['telecom_change_yn']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

)
dt_clf = DecisionTreeClassifier(
    max_depth=8,            
    min_samples_split=50,  
    min_samples_leaf=20,   
    criterion='entropy',     
)


dt_clf.fit(X_train, y_train)


dt_train_score = dt_clf.score(X_train, y_train)
dt_test_score = dt_clf.score(X_test, y_test)

print(f"DT 훈련 점수: {dt_train_score:.4f}")
print(f"DT 테스트 점수: {dt_test_score:.4f}")
print("-" * 30)
print(classification_report(y_test, dt_clf.predict(X_test)))

# 시각화
plt.figure(figsize=(20, 10))
plot_tree(dt_clf, 
          feature_names=X.columns, 
          class_names=['Stay', 'Leave'], 
          filled=True, 
          fontsize=10)
plt.show()
```
