### 통신사 바꾼 횟수 (아이디별)

```python
import matplotlib.pyplot as plt
import pandas as pd
telecom_df = pd.read_csv('../../data/processed/telecom.csv')
telecom_df = telecom_df[(telecom_df['telecom_change_yn']==1)]
# 아이디별 9년동안 통신사 바꾼횟수
telecom_df_change = telecom_df['id'].value_counts()
# 그 횟수의 갯수
telecom_df_change = telecom_df_change.value_counts()
# 한 번도 안바꾼 사람 추가
telecom_df_change[0] = 23226
telecom_df_change = telecom_df_change.sort_index(ascending=True)
telecom_df_change
```

### 통신사 바꾼 횟수(통신사별)

```python
telecom_df = pd.read_csv('../../data/processed/telecom.csv')
telecom_df_change = telecom_df['telecom'].value_counts()
print(telecom_df_change)
telecom_df_change_y = telecom_df[telecom_df['telecom_change_yn']==1]
telecom_df_change2 = telecom_df_change_y['telecom_change'].value_counts()
print(telecom_df_change2)

print('이탈자 비율', (telecom_df_change2/telecom_df_change)*100)
```
