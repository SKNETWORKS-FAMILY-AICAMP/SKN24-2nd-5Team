# 2. 데이터프레임 병합

```python
# 1. 9개년 데이터 묶기
data_frames = [p_17, p_18, p_19, p_20, p_21, p_22, p_23, p_24, p_25]

# 2. 하나로 다 합치기
df_all = pd.concat(data_frames, axis=0, ignore_index=True)
df_all.head()
```

```python
# 3. id 개수
id_counts = df_all['id'].value_counts()

# 4. 9개년 모두 응답한(즉, 9번 등장한) pid만 골라내기
valid_ids = id_counts[id_counts == 9].index

# 5. df_all에서, 9번 모두 등장한 id를 가진 행만 추출
df_balanced = df_all[df_all['id'].isin(valid_ids)].copy()

# 6. 사람별(id), 연도별(year)로 정리
df_tel = df_balanced.sort_values(by=['id', 'year']).reset_index(drop=True)
df_tel.head()
```

# 3. 결측치 처리 및 True/False 변환

```python
df_tel['gender'] = df_tel['gender'].replace({1: 0, 2: 1})
df_tel['job'] = df_tel['job'].replace({1: 1, 2: 0})
```

```python
df_tel['phone_usage_per_m'].value_counts()
```

```python
df_tel['mobile_bundle'].value_counts()
```

```python
df_tel['telecom'].value_counts()
```

```python
# 1. 빈칸(' ', '')이나 공백을 결측치(NaN)로 변환
df_tel = df_tel.replace(r'^\s*$', np.nan, regex=True)

# 2. 결측치가 하나라도 있는 행의 'id'를 중복 없이 추출
bad_user_ids = df_tel[df_tel.isna().any(axis=1)]['id'].unique()         # 전체 유저 중 결측치를 한 번이라도 낸 유저: 1116명

# 3. 전체 데이터셋에서 블랙리스트에 포함된 id를 가진 모든 행을 삭제
df_tel_clean = df_tel[~df_tel['id'].isin(bad_user_ids)].copy()          # 필터링하고 남은 것: 34776행

# 4. 쪼개져 있던 값들(문자 '1', 숫자 1 등)을 정수형으로 깔끔하게 통일
cols_to_fix = ['telecom', 'mobile_bundle', 'phone_usage_per_m']
df_tel_clean[cols_to_fix] = df_tel_clean[cols_to_fix].astype(int)
df_tel = df_tel_clean.copy()

# 6. 최종 확인
df_tel.info()
```

```python
df_tel['mobile_bundle'] = df_tel['mobile_bundle'].replace({1: 1, 2: 0})
```

# 4. 이탈여부 열 추가

### 4-1. Target data
- 작년 통신사와 현재 통신사가 달라지는 경우 이탈로 봄

```python
# 1. 정렬 및 변수 세팅
df_tel = df_tel.sort_values(by=['id', 'year']).reset_index(drop=True)
carrier_col = 'telecom' 

# 통신사 이름도 영어로 맞춤
telecom_map = {
    1: 'SKT', 2: 'KT', 3: 'LGU', 4: 'MVNO',
}

# 2. 작년 통신사 끌어오기
df_tel['prev_telecom'] = df_tel.groupby('id')[carrier_col].shift(1)

# 3. 2017년(첫 해) NaN 해결: 빈칸을 '올해 통신사'로 채워서 유지(0)로 간주
df_tel['prev_telecom'] = df_tel['prev_telecom'].fillna(df_tel[carrier_col]).astype(int)

# 4. 이탈 여부 계산
df_tel['telcom_changed'] = np.where(df_tel['prev_telecom'] != df_tel[carrier_col], 1, 0)

# 5. EDA용 영어 상태 변수(churn_status) 만들기
df_tel['prev_name'] = df_tel['prev_telecom'].map(telecom_map)
df_tel['curr_name'] = df_tel[carrier_col].map(telecom_map)

# 유지 = "SKT_Stay", 이탈 = "SKT_to_KT" 형태로 텍스트 생성
df_tel['churn_status_1'] = np.where(
    df_tel['telcom_changed'] == 0,
    df_tel['prev_name'] + '_Stay',  
    df_tel['prev_name'] + '_to_' + df_tel['curr_name']
)

# 6. 임시로 만든 이름 열은 지우고 마무리
df_tel = df_tel.drop(['prev_name', 'curr_name'], axis=1).copy()
df_tel['telcom_changed'] = df_tel['telcom_changed'].fillna(0).astype(int)
df_tel.head()
```

### 4-2. 이탈 방향에 따른 열 추가 (for EDA)

```python
# 1. 딕셔너리
status_dict = {
    'SKT_Stay': 1, 'KT_Stay': 2, 'LGU_Stay': 3, 'MVNO_Stay': 4,
    'SKT_to_KT': 12, 'SKT_to_LGU': 13, 'SKT_to_MVNO': 14,
    'KT_to_SKT': 21, 'KT_to_LGU': 23, 'KT_to_MVNO': 24,
    'LGU_to_SKT': 31, 'LGU_to_KT': 32, 'LGU_to_MVNO': 34, 
    'MVNO_to_SKT': 41, 'MVNO_to_KT': 42, 'MVNO_to_LGU': 43 
}

# 2. 매핑과 동시에 정수형(int)으로 변환하기!
df_tel['churn_status_num_1'] = df_tel['churn_status_1'].map(status_dict).astype(int)

# 3. prev_telecom 열 지우고 원본 데이터(df_tel)에 덮어쓰기
df_tel = df_tel.drop(['prev_telecom'], axis=1).copy()
```

```python
# 1. 요약용 딕셔너리 생성
simple_dict_1 = {
    'SKT_Stay': 'SKT_stay', 'KT_Stay': 'KT_stay', 'LGU_Stay': 'LGU_stay', 'MVNO_Stay': 'MVNO_stay',
    'SKT_to_KT': 'SKT_churned', 'SKT_to_LGU': 'SKT_churned', 'SKT_to_MVNO': 'SKT_churned',
    'KT_to_SKT': 'KT_churned', 'KT_to_LGU': 'KT_churned', 'KT_to_MVNO': 'KT_churned',
    'LGU_to_SKT': 'LGU_churned', 'LGU_to_KT': 'LGU_churned', 'LGU_to_MVNO': 'LGU_churned',
    'MVNO_to_SKT': 'MVNO_churned', 'MVNO_to_KT': 'MVNO_churned', 'MVNO_to_LGU': 'MVNO_churned'
}

# 2. 새로운 열 추가
df_tel['churn_status_2'] = df_tel['churn_status_1'].map(simple_dict_1)
```

```python
# 1. 요약된 값을 숫자로 바꾸는 새 딕셔너리
final_num_dict = {
    'SKT_stay': 1, 'KT_stay': 2, 'LGU_stay': 3, 'MVNO_stay': 4,
    'SKT_churned': 10, 'KT_churned': 20, 'LGU_churned': 30, 'MVNO_churned': 40
}

# 2. churn_status_2(문자열)를 기반으로 churn_status_num_2(숫자) 생성
df_tel['churn_status_num_2'] = df_tel['churn_status_2'].map(final_num_dict).astype(int)
```
